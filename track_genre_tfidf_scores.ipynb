# idf scores

```python
track_genres_df = pd.read_csv("feature_engineering/data/track_genres.csv", sep=";", header=None)
track_genres_df.columns = ["track_id", "genres"]
track_genres_df = track_genres_df[track_genres_df["genres"] != "[]"]
track_genres_df.set_index("track_id", inplace=True)
track_genres_df["genres"] = track_genres_df["genres"].apply(lambda r: ast.literal_eval(r))
track_genres_df.head()
```

```python
relevant_tracks = pd.read_csv("feature_engineering/data/lowms_events_nondominating.csv", sep=";")["track_id"].unique().tolist()
relevant_tracks
```

```python
len(relevant_tracks)
```

```python
user_track_df = pd.read_csv("feature_engineering/data/user_track.csv", sep=";")
rel_tracks = user_track_df["track_id"].unique()
```

```python
track_genres_df = track_genres_df.loc[relevant_tracks].dropna()
```

```python
n_documents = len(track_genres_df)
n_documents
```

```python
unique_genres = set()
for genres in track_genres_df["genres"]:
    unique_genres = unique_genres.union(genres)
len(unique_genres)
```

```python
track_genres_df["genres"] = track_genres_df["genres"].apply(lambda r: Counter(set(r)))
track_genres_df.head()
```

```python
n_rel_docs = Counter(dict.fromkeys(unique_genres, 0))
for _, counts in track_genres_df.iterrows():
    n_rel_docs = Counter(n_rel_docs + counts.values[0])
n_rel_docs
```

```python
idf_scores = {genre: np.log10(n_documents / n_rel) for genre, n_rel in dict(n_rel_docs).items()}
idf_scores_df = pd.DataFrame.from_dict(idf_scores, orient="index", columns=["idf"]).sort_values(by="idf", ascending=True)
idf_scores_df.head(16)
```

```python
%matplotlib qt
#plt.figure(figsize=(8, 6))
plt.plot(list(range(0, 100)), idf_scores_df.iloc[:100], "-o")
plt.xlabel("Top-100 genres")
plt.ylabel("Genre IDF-score")
plt.axhline(y=0.9, linestyle="--", color="black", label="Lower bound (0.90)")
plt.legend()
plt.grid(False)
plt.show()
```

```python
idf_scores_df.to_csv("track_genres_idf_dist.csv", sep=";")
```

# idf per cluster

```python
track_cluster_df = pd.read_csv("clustering/track_to_cluster.csv", sep=";")
track_cluster_df.head()
```

```python
track_cluster_df.groupby(by="cluster").count()
```

```python
track_cluster_df.set_index("track_id", inplace=True)
```

```python
tf_idf_scores_df = pd.DataFrame()
for cluster, group in track_genres_df.merge(track_cluster_df, left_index=True, right_index=True).groupby(by="cluster"):
    term_freqs = Counter(dict.fromkeys(unique_genres, 0))
    for _, counts in group.iterrows():
        term_freqs = Counter(term_freqs + Counter(counts.values[0]))
    tf_idf = Counter({genre: idf_scores[genre] * term_freqs[genre] for genre in unique_genres})
    tf_idf_df = pd.DataFrame.from_dict(tf_idf, orient="index", columns=["tf-idf"]).sort_values(by="tf-idf", ascending=False)
    tf_idf_df.index.name = "genre"
    tf_idf_df["cluster"] = cluster
    tf_idf_scores_df = tf_idf_scores_df.append(tf_idf_df)
```

```python
clean_idf_scores_df = tf_idf_scores_df.copy()
clean_idf_scores_df.reset_index(inplace=True)
clean_idf_scores_df.set_index("cluster", inplace=True)
clean_idf_scores_df = clean_idf_scores_df[~clean_idf_scores_df["genre"].isin(["rock", "pop", "electronic", "metal", "alternativerock", "indierock"])]
```
